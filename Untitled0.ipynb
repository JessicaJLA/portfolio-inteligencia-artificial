```python
import pandas as pd

# Criando um dataset fictício
dados = {
    'idade': [25, 34, 45, 23, 54, 31, 40, 29, 60, 48],
    'categoria': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A', 'C'],
    'gasto': [200, 350, 400, 150, 500, 300, 220, 410, 600, 280]
}

df = pd.DataFrame(dados)

# Salvar CSV
df.to_csv('dados_clientes.csv', index=False)

df
```

```python
# Informações gerais
df.info()

# Estatísticas descritivas
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribuição de idade
sns.histplot(df['idade'], kde=True)
plt.title('Distribuição de Idade dos Clientes')
plt.show()

# Gasto por categoria
sns.boxplot(x='categoria', y='gasto', data=df)
plt.title('Gasto por Categoria')
plt.show()
```

```python
# Gasto médio por categoria
gasto_medio = df.groupby('categoria')['gasto'].mean()
gasto_medio
```

## 📊 Principais Insights

- A categoria A apresentou maior gasto médio.
- Observa-se que clientes com idade mais elevada tendem a gastar mais.
- A análise demonstra como dados simples podem gerar informações relevantes para tomada de decisão.
